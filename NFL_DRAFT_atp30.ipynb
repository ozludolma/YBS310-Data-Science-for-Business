```python
drafted = NFL[NFL['Draft Status'] == 'Drafted']
group = NFL.groupby(['college', 'Position'])['Draft Status'].value_counts().unstack().fillna(0)
group['draft_rate'] = group['Drafted'] / (group['Drafted'] + group['Undrafted'])
```

```python
min_players = 5
college_position_summary = group[group['Drafted'] + group['Undrafted'] >= min_players]
top_colleges_by_position = college_position_summary.groupby('Position')['draft_rate'].idxmax()

print("Best college for each position (by draft rate, min 5 players):")
for pos, idx in top_colleges_by_position.items():
    college, _ = idx
    rate = college_position_summary.loc[idx, 'draft_rate']
    total = int(college_position_summary.loc[idx, 'Drafted'] + college_position_summary.loc[idx, 'Undrafted'])
    print(f"{pos}: {college} ({rate:.2%} draft rate, {total} players)")
```

```python
pivot = college_position_summary['draft_rate'].unstack()
plt.figure(figsize=(12, 8))
sns.heatmap(pivot, annot=True, fmt=".1%", cmap='plasma')
plt.title('Draft Rate by College and Position')
plt.ylabel('College')
plt.xlabel('Position')
plt.show()
```

```python
first_round = drafted[drafted['Pick Round'] == '1st']
group_first = NFL.groupby(['college', 'Position'])['Draft Status'].count().unstack().fillna(0)
first_group = first_round.groupby(['college', 'Position'])['Draft Status'].count().unstack().fillna(0)
first_rate = (first_group / group_first).fillna(0)
```

```python
first_college_summary = group_first[group_first.sum(axis=1) >= min_players]
top_colleges_first_by_position = first_rate.idxmax()

print("Best college for each position to go 1st round (among colleges with at least 5 draftees):")
for pos in first_rate.columns:
    college = top_colleges_first_by_position[pos]
    rate = first_rate.loc[college, pos] if college in first_rate.index else 0
    total = int(group_first.loc[college, pos]) if college in group_first.index else 0
    if total >= min_players:
        print(f"{pos}: {college} ({rate:.2%} 1st round rate, {total} drafted players)")
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
ct = ColumnTransformer([("onehot", OneHotEncoder(), categorical_data)])
NFL_cont = NFL.drop(categorical_data, axis=1)
NFL_categorical = ct.fit_transform(NFL)
NFL_onehot = pd.concat(
    [NFL_cont.reset_index(drop=True), pd.DataFrame(NFL_categorical.toarray(), columns=ct.get_feature_names_out())],
    axis=1
)

print("After one-hot encoding, missing values per column:")
print(NFL_onehot.isna().sum())
NFL_onehot.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
```

```python
X = NFL_onehot.drop(NFL_onehot.columns[-1], axis=1)
y = NFL_onehot.iloc[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
print("Train/test shapes:", np.shape(X_train), np.shape(y_train), np.shape(X_test), np.shape(y_test))
```

```python
log_reg_model = LogisticRegression(max_iter=1000).fit(X_train, y_train)
```

```python
y_pred = log_reg_model.predict(X_test)
cm_test = confusion_matrix(y_test, y_pred, labels=log_reg_model.classes_)

disp = ConfusionMatrixDisplay(confusion_matrix=cm_test, display_labels=log_reg_model.classes_)
disp.plot(cmap='Paired') 
plt.title('Confusion Matrix: Logistic Regression Draft Prediction')
plt.show()
```

```python
print("Model Accuracy:", log_reg_model.score(X_test, y_test))
print(classification_report(y_test, y_pred, target_names=['drafted', 'unDrafted']))
```

Modeling using categorical bayes 

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.naive_bayes import CategoricalNB
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

```python
NFL = pd.read_excel("C:/Users/HP/Desktop/NFL DRAFT DATABASE02.0.xlsx")
```

```python
print("Original data sample:")
print(NFL.head())
print(f"Shape: {NFL.shape}")
```

```python
cols_to_drop = ['Year', '40yard dash', 'Vertical Jump', 'Draft Status', 'Drafting team', 'Pick number']
cols_to_drop = [col for col in cols_to_drop if col in NFL.columns]
NFL_nb = NFL.drop(columns=cols_to_drop)
print("\nData after dropping columns:")
print(NFL_nb.head())
print(f"Shape: {NFL_nb.shape}")
```

```python
NFL_nb['first_round'] = NFL_nb['Pick Round'].apply(lambda x: 1 if str(x).strip().lower() == '1st' else 0)
print("\nSample of Pick Round and first_round:")
print(NFL_nb[['Pick Round', 'first_round']].head(10))
```

```python
X = NFL_nb.drop(columns=['Pick Round', 'first_round'])
y = NFL_nb['first_round']
print("\nTarget value counts:")
print(y.value_counts())
```

```python
cat_features = X.select_dtypes(include="object").columns
print("\nCategorical columns:", cat_features.tolist())
```

```python
ct_ord = ColumnTransformer([("ord", OrdinalEncoder(), cat_features)], remainder='passthrough')
X_ord = ct_ord.fit_transform(X)
X_ord = pd.DataFrame(X_ord, columns=ct_ord.get_feature_names_out())
print("\nOrdinal encoded feature sample:")
print(X_ord.head())
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_ord, y, test_size=0.2, random_state=42, stratify=y)
```

```python
classifier = CategoricalNB(alpha=0, force_alpha=True)
classifier.fit(X_train, y_train.ravel())
```

```python
y_pred = classifier.predict(X_test)
print("\nClassification report for CategoricalNB (1st round prediction):")
print(classification_report(y_test, y_pred, labels=[0, 1], target_names=['Not 1st Round', '1st Round']))
```

```python
y_pred = classifier.predict(X_test)
print("\nClassification report for CategoricalNB (1st round prediction):")
print(classification_report(y_test, y_pred, labels=[0, 1], target_names=['Not 1st Round', '1st Round']))
```

```python
cm = confusion_matrix(y_test, y_pred, labels=[0, 1])
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Not 1st Round', '1st Round'])
disp.plot(cmap='YlOrRd')
plt.title('Confusion Matrix: CategoricalNB First Round Prediction')
plt.show()
```
